Create a library catalog system using OOP. The system should allow users to add books to the catalog, search for books by title or author, and remove books from the catalog.
Requirements:
1.Each book should have a title, author, and unique identifier
2.The system should provide the following functionality:
    ●Adding a book to the catalog.
    ●Searching for books by title or author.
    ●Removing a book from the catalog.
    ●Displaying all books in the catalog.
3.Do the exception handling4


```python
class book:
    def __init__(self,title,author,unique):
        self.title = title
        self.author = author
        self.unique = unique

    def __str__(self):
        return f"judul {self.title} author {self.author} unique {self.unique}"
    
class library:
    def __init__(self):
        self.catalog=[]
    
    def tambah(self,isi):
        self.catalog.append(isi)
        print(f"{isi.title} by {isi.author} berhasil ditambah.")
    
    def cari(self,search):
        found = []
        for caribuku in self.catalog:
            if search.lower() in caribuku.title.lower() or search.lower() in caribuku.author.lower():
                found.append(caribuku)
                print(caribuku)
            else:
                print("buku tidak ada")


    def hapus(self,title):
        found = []
        for hapus in self.catalog:
            if hapus.title.lower() == title.lower():
                self.catalog.remove(hapus)
                print(f"{title} berhasil dihapus")
                found = True
                break
        if not found:
            print(f"{title} tidak ada")

    def tampil(self):
        if self.catalog:
            print("isi perpus")
            for isi in self.catalog:
                print (isi)
        else:
            print("tidak ada isi")

def main():
    my_library=library()
    while True:
        print("Menu")
        print("1. tambah buku")
        print("2. cari buku")
        print("3. haous buku")
        print("4. tampilkan buku")
        print("5. exit")
        try:
            pilih=input("masukan menu")
            if pilih == '1':
                title = input("masukan judul")
                author = input ("masukan author")
                unique = input("masukan unique ID")
                my_library.tambah(book(title,author,unique))

            elif pilih =='2':
                search=input("masukan yg ingin dicari")
                print(f"pencarian buku: {search}")
                my_library.cari(search)

            elif pilih == '3':
                remove=input("masukan yg mau dihapus")
                my_library.hapus(remove)

            elif pilih == '4':
                my_library.tampil()

            elif pilih=='exit' or pilih=='5':
                break
            
            else:
                print("invalid pilihh")

        except ValueError:
            print("invalid pilih")    

if __name__=='__main__':
    main()

```
